```python
import numpy as np
import matplotlib.pyplot as plt

# [Math, Physics, Chemistry, English, CS]
data = np.array([
    [1, 88, 92, 85, 91, 89],
    [2, 76, 81, 79, 77, 80],
    [3, 93, 89, 94, 90, 96],
    [4, 65, 70, 60, 68, 72],
    [5, 85, 87, 83, 86, 84],
])
```

```python
# Mean or average
print("Subject wise mean: ",np.mean(data[:,1:] , axis=0))
print("Student wise mean: ",np.mean(data[:,1:] , axis=1))
```

```python
# Standard Deviation
print("Standard Deviation per subject: ",np.std(data[:,1:], axis=0))
```

```python
# Highest marks
total_marks = np.sum(data[:,1:], axis=1)
print("Total marks per student", total_marks)
print(f"Maximum Marks scored by Student-{np.argmax(total_marks)+ 1} with total marks {np.max(total_marks)}")
```

```python


subject_names = ['Math', 'Physics', 'Chemistry', 'English', 'CS']
subject_means = np.mean(data[:,1:], axis=0)

plt.figure(figsize=(10,6))
plt.bar(subject_names, subject_means, color='teal')
plt.title('Average Marks per Subject')
plt.xlabel('subject')
plt.ylabel('Marks')
plt.grid(True)
plt.show()
```
